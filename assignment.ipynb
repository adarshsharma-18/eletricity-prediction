```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

# Example data
X = np.array([[1, 2], [1, 4], [1, 0], [10, 2], [10, 4], [10, 0]])

# Initialize KMeans with 2 clusters
kmeans = KMeans(n_clusters=2, random_state=0).fit(X)

# Print the cluster centroids and labels
print("Cluster centers:", kmeans.cluster_centers_)
print("Labels:", kmeans.labels_)

# Plotting the clusters
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='black', marker='x')
plt.show()
```

```python
from sklearn.mixture import GaussianMixture

# Initialize GMM with 2 components
gmm = GaussianMixture(n_components=2, random_state=0).fit(X)

# Predict the labels (cluster assignment)
labels = gmm.predict(X)

# Print the parameters (means and covariances of the Gaussians)
print("Means:", gmm.means_)
print("Covariances:", gmm.covariances_)

# Plotting the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow')
plt.scatter(gmm.means_[:, 0], gmm.means_[:, 1], s=300, c='black', marker='x')
plt.show()
```

```python

from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

# Create Agglomerative Clustering
agg = AgglomerativeClustering(n_clusters=2)
labels = agg.fit_predict(X)

# Plot clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow')
plt.show()

# Plot dendrogram
Z = linkage(X, 'ward')
dendrogram(Z)
plt.show()

```

```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Sample 3D data
X = np.array([[2.5, 2.4, 1.5],
              [0.5, 0.7, 0.3],
              [2.2, 2.9, 1.6],
              [1.9, 2.2, 1.3],
              [3.1, 3.0, 2.0],
              [2.3, 2.7, 1.5]])

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Create PCA object
pca = PCA(n_components=2)

# Fit and transform the data
X_pca = pca.fit_transform(X_scaled)

# Print the explained variance ratio
print("Explained Variance Ratio:", pca.explained_variance_ratio_)

# Plotting the PCA results
plt.scatter(X_pca[:, 0], X_pca[:, 1], cmap='rainbow')
plt.title("2D PCA projection of the data")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.grid()
plt.show()
```
